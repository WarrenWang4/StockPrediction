```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

color = sns.color_palette()
sns.set_style('darkgrid')
```

```python
fang = pd.read_csv(os.getcwd() +"/data.tmp/processed/fang.csv", low_memory=False)
fang
```

```python
apple = fang[fang['Name'] == 'AAPL']
netflix = fang[fang['Name'] == 'NFLX']
amazon = fang[fang['Name'] == 'AMZN']
facebook = fang[fang['Name'] == 'FB']
google = fang[fang['Name'] == 'GOOGL']

list = [apple,netflix,amazon,facebook,google]

for df in list:
    df.set_index('Date', inplace=True)
    df.index.sort_values()

print(apple)
print(netflix)
print(amazon)
print(facebook)
print(google)
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="Date", y="Close",legend = 'full' , data=apple)
```
